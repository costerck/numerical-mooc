```python
from matplotlib import pyplot
import numpy
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16

n=100000
minx = 0.0
maxx = 0.01*n
dx = (maxx-minx)/(n-1)

x = numpy.linspace(minx, maxx, n)
xprim = numpy.linspace(minx+2*dx, maxx-2*dx, n-4)
y = x**4
u=numpy.zeros(n-4)

for i in range(2, 98):
    u[i-2] = (y[i+2]-4*y[i+1]+6*y[i]-4*y[i-1]+y[i-2])/dx/dx/dx/dx

pyplot.plot(x, y)
pyplot.plot(xprim, u)

print(numpy.max(u[:]))
```
